## Часть 1 ##

### Подготовка датасета. ###

```python
# Настроим токенизатор
from transformers import AutoTokenizer

model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Загрузка датасета
from datasets import load_dataset

raw_dataset = load_dataset("stanfordnlp/imdb") # Возьмем датасет stanfordnlp/imdb
print(raw_dataset)
print(raw_dataset["train"][0])
```

```python
# Процесс токенизации
def tokenize_batch(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=128,
    )

tokenized_dataset = raw_dataset.map(tokenize_batch, batched=True)
tokenized_dataset = tokenized_dataset.remove_columns(["text"])
tokenized_dataset = tokenized_dataset.rename_column("label", "labels")
tokenized_dataset.set_format("torch")
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(
    tokenized_dataset["train"], shuffle=True, batch_size=128
)
eval_dataloader = DataLoader(
    tokenized_dataset["test"], batch_size=128
)
```

### Загрузка модели и настройко компонентов обучения 

```python
import torch

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
device
```

```python
# Настроим классификатор
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2).to(device)
```

```python
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr=5e-5)
```

```python
from transformers import get_scheduler

num_epochs = 3
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps,
)
print(num_training_steps)
```

### Обучение

```python
from tqdm.auto import tqdm

progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
```

```python
import evaluate
accuracy = evaluate.load("accuracy")
model.eval()
for batch in eval_dataloader:
    batch = {k: v.to(device) for k, v in batch.items()}
    with torch.no_grad():
        outputs = model(**batch)

    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1)
    accuracy.add_batch(predictions=predictions, references=batch["labels"])

print(accuracy.compute())
```

Итого, на тестовой выборке показатель accuracy составил 0.88932 для модели bert-base-uncased, обученной на 3-х эпохах

## Часть 2

### Классификатор с кастомной головой

Напишем класс для классификатор с кастомной головой на базе bert-base-uncased, добавив один линейный слой

```python
import torch
import torch.nn as nn
from transformers import AutoConfig, AutoModel
from torch.nn import CrossEntropyLoss

class TransformerForSequenceClassification(nn.Module):
    def __init__(self, config_name: str, num_labels: int):
        super().__init__()
        config = AutoConfig.from_pretrained(config_name, num_labels=num_labels)
        self.encoder = AutoModel.from_pretrained(config_name, config=config)
        self.dropout = nn.Dropout(config.hidden_dropout_prob)
        self.classifier = nn.Linear(config.hidden_size, num_labels)
        self.loss_fct = CrossEntropyLoss()

    def forward(self, input_ids, attention_mask=None, token_type_ids=None, labels=None):
        outputs = self.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            return_dict=True,
        )
        cls_state = outputs.last_hidden_state[:, 0, :]
        cls_state = self.dropout(cls_state)
        logits = self.classifier(cls_state)

        loss = None
        if labels is not None:
            loss = self.loss_fct(logits.view(-1, self.classifier.out_features), labels.view(-1))

        return (loss, logits) if loss is not None else logits
```

```python
model = TransformerForSequenceClassification(model_name, 2).to(device)
```

### Обучение

Запустим обучение с помощью класса Trainer, позволяющего упростить многие процессы

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=128,
    per_device_eval_batch_size=128,
    num_train_epochs=3,
    weight_decay=0.01,
    load_best_model_at_end=True,
    report_to="none",
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    processing_class=tokenizer,
)
```

```python
trainer.train()
```

```python
import evaluate
accuracy = evaluate.load("accuracy")
model.eval()
for batch in eval_dataloader:
    batch = {k: v.to(device) for k, v in batch.items()}
    with torch.no_grad():
        outputs = model(**batch)

    logits = outputs[1]
    predictions = torch.argmax(logits, dim=-1)
    accuracy.add_batch(predictions=predictions, references=batch["labels"])

print(accuracy.compute())
```

Итого, на тестовой выборке показатель accuracy составил 0.88696 для модели кастомной головой, обученной на 3-х эпохах. Показатель стал немного ниже в сравнении с моделью из AutoModelForSequenceClassification (с показателем 0.88932), но не значительно. Скорее всего это связано с тем, что в данном классе используется более сложная архитектура "головы", возможно с несколькими линейными слоями или нелинейностями.

## Задание 3 (XLNET)



### Подготовка датасета
Для работы с датасетом, надо его предобработать, токензировав лейблы

```python
# Настроим токенизатор
from transformers import AutoTokenizer

# model_name = "bert-base-cased"
model_name = "xlnet-base-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Загрузка датасета
from datasets import load_dataset

raw_dataset = load_dataset("eriktks/conll2003")
print(raw_dataset)
print(raw_dataset["train"][0])
```

```python
label_all_tokens = True
task = 'ner'
label_list = raw_dataset["train"].features[f"{task}_tags"].feature.names
label_list
```

```python
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True)

    labels = []
    for i, label in enumerate(examples[f"{task}_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx])
            else:
                label_ids.append(label[word_idx] if label_all_tokens else -100)
            previous_word_idx = word_idx

        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs

tokenized_dataset = raw_dataset.map(
    tokenize_and_align_labels,
    batched=True,
)
```

### Настройка модели и параметров обучения

```python
from transformers import AutoModelForTokenClassification

model = AutoModelForTokenClassification.from_pretrained(model_name, num_labels=len(label_list))
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results_3",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    weight_decay=0.01,
    load_best_model_at_end=True,
    report_to=["tensorboard"],
)
```

```python
from transformers import DataCollatorForTokenClassification

data_collator = DataCollatorForTokenClassification(tokenizer)
```

```python
import evaluate

metric = evaluate.load("seqeval")
```

```python
import numpy as np

def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)

    # Remove ignored index (special tokens)
    true_predictions = [
        [label_list[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [label_list[l] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]

    results = metric.compute(predictions=true_predictions, references=true_labels)
    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }
```

### Обучение

```python
from transformers import Trainer

trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```
